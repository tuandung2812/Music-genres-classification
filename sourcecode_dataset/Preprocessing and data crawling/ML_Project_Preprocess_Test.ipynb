```python
import librosa
import librosa.display
import csv
import os
import numpy as np
import skimage.io
```

# Extract log mel spectrogram (Splitted)

```python
os.makedirs('mel_spectrogram_test')
```

```python
for file_name in os.listdir('/content/drive/MyDrive/separate_test'):
  if file_name not in os.listdir('/content/mel_spectrogram_test') and not file_name.endswith('.mf'):
    os.mkdir('/content/mel_spectrogram_test/' + file_name)
    for wav_file in os.listdir('/content/drive/MyDrive/separate_test/' + file_name):
      audio, sampling_rate = librosa.load('/content/drive/MyDrive/separate_test/' + file_name + '/' + wav_file, duration=40)
      for step in range(0, len(audio) - len(audio)//10, len(audio)//10):
        mel_spec = librosa.feature.melspectrogram(audio[step:step+len(audio)//10])
        mel_spec = (mel_spec - np.min(mel_spec)) / np.max(mel_spec) ** 3
        try:
          skimage.io.imsave('mel_spectrogram_test/' + file_name + '/' + wav_file[:-4] + '_' + str(step) + '.png', np.flip(librosa.power_to_db(mel_spec, ref=np.max), axis=0))
        except:
          print(wav_file[:-4] + '_' + str(step) + '.png')
```

Save spectrogram as images & stored in tar file for later uses

```python
import tarfile
import os.path
 
def make_tarfile(output_filename, source_dir):
    with tarfile.open(output_filename, "w:gz") as tar:
        tar.add(source_dir, arcname=os.path.basename(source_dir))
 
make_tarfile('mel_spec_test.tar.gz', '/content/mel_spectrogram_test')
```
